## Exploratory Data Analysis

```python
# Data Visualization.
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(10,3)})
ax = sns.countplot(data = data,x = 'subject',hue='subject')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(10,3)})
sales_subject = data.groupby(['subject'],as_index=False)['price'].sum().sort_values(by = 'price',ascending=False)
sns.barplot(data = sales_subject, x = 'subject',y='price',hue='subject')
```

*From above graphs we see that most of the sold courses are from Web development, Business Finance and Musicial Instruments subjects.*

```python
sns.set(rc={'figure.figsize':(10,3)})
ax = sns.countplot(data = data, x = 'is_paid',hue='is_paid')

for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graph we see that most sold courses are paid.*

```python
sns.set(rc={'figure.figsize':(10,3)})
ax = sns.countplot(data = data,x = 'Year',hue='Year')

for bars in ax.containers:
    ax.bar_label(bars)
```

*We can see that most of sold courses in the year of 2016, 2015 and 2017.*

```python
subject_subscribe = data.groupby(['subject'],as_index=False)['num_subscribers'].sum().sort_values(by='num_subscribers',ascending=False)

sns.set(rc={'figure.figsize':(10,3)})
sns.barplot(data = subject_subscribe, x = 'subject',y = 'num_subscribers', hue ='subject')
```

*From graph we can see that most of the subscribers are from the web development subject*

```python
sns.set(rc={'figure.figsize':(10,6)})
ax = sns.countplot(data = data, x ='subject',hue='Year')

for bars in ax.containers:
    ax.bar_label(bars)
```

## Conclusion
#### 1.Popular Subject: Popular subjects among users are Web Development,Business Finance, and Musicial Instruments.
#### 2.Monetization Strategy: Paid Courses significantly outnumber free ones in terms of sales.
#### 3.Target Audience: Preferences for courses related to Web Development among subscribers.
#### 4.Sales Trends: Notable increases in course sales occurred in 2015,2016,and 2017.
